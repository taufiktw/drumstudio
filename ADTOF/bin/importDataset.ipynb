```python
"""Import the dataset"""
import tensorflow as tf
import matplotlib.pyplot as plt

dataset = tf.data.Dataset.load("../../datasets/export adtofv2/adtof_rgw_test") # Modify the path
```

```python
"""plot one sample"""

for input, target in dataset:
    fig, ax = plt.subplots(2)

    ax[0].set_title("Input (400 frames, 84 bins, 1 channel)")
    ax[0].imshow(input["x"][:, :, 0].numpy().T)
    ax[0].set_xlabel("Frames (hop size: 10ms)")
    ax[0].set_ylabel("Frequency bins")

    ax[1].set_title("Target (400 frames, 5 classes)")
    ax[1].imshow(target.numpy().T, aspect=10, interpolation="nearest")
    ax[1].set_xlabel("Frames (hop size: 10ms)")
    ax[1].set_ylabel("Classes")
    ax[1].set_yticks([0, 1, 2, 3, 4], ("BD", "SD", "TT", "HH", "CY+RD"))
    fig.tight_layout()
    plt.show()
    break

```
